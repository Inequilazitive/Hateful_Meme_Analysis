# SCRIPT 1

```python
import os
import cv2
import torch
import pandas as pd
import numpy as np
from PIL import Image
from tqdm import tqdm
from yolov5.utils.general import non_max_suppression, scale_boxes
from torchvision import transforms
import keras_ocr
from yolov5.models.experimental import attempt_load

image_folder = "hateful_memes/img"
weights_path = "yolov5s.pt"
model = attempt_load(weights_path)
model.eval()
print(model)

used_images_df = pd.read_csv("used_images_list.csv")
pipeline = keras_ocr.pipeline.Pipeline()
output_folder = "processed_images"
os.makedirs(output_folder, exist_ok=True)
comparison_results_file = "comparison_results.csv"
comparison_results = []

to_tensor = transforms.Compose([
    transforms.ToTensor(),
    transforms.Lambda(lambda x: x[:3, :, :]),
])

for idx, row in tqdm(used_images_df.iterrows(), total=len(used_images_df)):
    filename = row["Image"]
    img_path = os.path.join(image_folder, filename)
    img_original = Image.open(img_path)
    img_original = img_original.convert('RGB')
    img_original = img_original.resize((640, 640))
    img_tensor = to_tensor(img_original).unsqueeze(0)
    results_original = model(img_tensor)
    results_np_original = non_max_suppression(results_original, conf_thres=0.5, iou_thres=0.5)[0]
    prediction_groups = pipeline.recognize([img_path])
    mask = np.zeros(img_original.size[::-1], dtype="uint8")
    for box in prediction_groups[0]:
        box_np = np.array(box[1], dtype=np.int32)
        cv2.fillPoly(mask, [box_np], 255)
    img_without_captions = Image.fromarray(np.array(img_original) * (1 - mask[:, :, None]))
    img_without_captions.save(os.path.join(output_folder, f"img_without_captions_{filename}"))
    img_tensor_without_captions = to_tensor(img_without_captions).unsqueeze(0)
    results_without_captions = model(img_tensor_without_captions)
    results_np_without_captions = non_max_suppression(results_without_captions, conf_thres=0.5, iou_thres=0.5)[0]
    comparison_results.append({
        "Sr. No": row["Sr. No"],
        "Image": filename,
        "NumObjects_Original": len(results_np_original) if results_np_original is not None else 0,
        "NumObjects_WithoutCaptions": len(results_np_without_captions) if results_np_without_captions is not None else 0,
    })

used_images_df = pd.DataFrame(used_images_df)
comparison_results_df = pd.DataFrame(comparison_results)
used_images_df.to_csv("used_images_list2.csv", index=False)
comparison_results_df.to_csv(comparison_results_file, index=False)
```

# SCRIPT 2

```python
import os
import cv2
#import tesseract
import pandas as pd
from tqdm import tqdm
from PIL import Image
from torchvision import transforms
from yolov5.models.experimental import attempt_load
from yolov5.utils.general import non_max_suppression, scale_boxes
import numpy as np
import keras_ocr


# Function to compute the gradients magnitude of the image
def compute_grad(image_gray, mode):
    dx = cv2.Sobel(image_gray, cv2.CV_64F, 1, 0, ksize=3)
    dy = cv2.Sobel(image_gray, cv2.CV_64F, 0, 1, ksize=3)
    mag = np.sqrt(dx**2 + dy**2)

    if mode == "double":
        dx = cv2.Sobel(mag, cv2.CV_64F, 1, 0, ksize=3)
        dy = cv2.Sobel(mag, cv2.CV_64F, 0, 1, ksize=3)
        magmag = np.sqrt(dx**2 + dy**2)
        return (magmag/magmag.max()*255).astype(np.uint8)

    return (mag/mag.max()*255).astype(np.uint8)

# Function to detect and draw bounding boxes around text
def detect_draw(pipeline, image_gray, viz):
    img = cv2.cvtColor(image_gray, cv2.COLOR_GRAY2RGB)
    b = pipeline.recognize([img])

    if viz:
        keras_ocr.tools.drawAnnotations(image=img, predictions=b[0])

    x_min = int(min([b[0][i][1][:,0].min() for i in range(len(b[0]))]))
    x_max = int(max([b[0][i][1][:,0].max() for i in range(len(b[0]))]))
    y_min = int(min([b[0][i][1][:,1].min() for i in range(len(b[0]))]))
    y_max = int(max([b[0][i][1][:,1].max() for i in range(len(b[0]))]))

    return (x_min, y_min, x_max, y_max)

# Function to remove text based on the detected bounding box
def remove(image_gray, bb, offset):
    mask = np.ones_like(image_gray) * 255
    cv2.rectangle(mask, (bb[0] - offset, bb[1] - offset), (bb[2] + offset, bb[3] + offset), (0, 0, 0), -1)
    masked_image = cv2.bitwise_and(image_gray, mask)

    return masked_image

# Function to remove captions based on the provided method
def remove_captions(image, text_detection_pipeline, offset):
    image_gray = cv2.cvtColor(np.array(image), cv2.COLOR_RGB2GRAY)
    mag = compute_grad(image_gray, "single")
    bb = detect_draw(text_detection_pipeline, mag, viz=False)
    cleaned_image_final = remove(image_gray, bb, offset)
    return cleaned_image_final

def perform_object_detection_with_text_removal(img_folder, weights_path, used_images_csv, cleaned_image_path, text_removal_offset=40):
    # Load YOLOv5 model with specified weights file
    model = attempt_load(weights_path)  # Use 'cpu' for device
    model.eval()

    used_images_df = pd.read_csv(used_images_csv)

    results_list = []

    to_tensor = transforms.Compose([
        transforms.ToTensor(),
        transforms.Lambda(lambda x: x[:3, :, :]),  # Keep only the first 3 channels (RGB)
    ])

    os.makedirs(cleaned_image_path, exist_ok=True)

    text_detection_pipeline = keras_ocr.pipeline.Pipeline()

    for idx, row in tqdm(used_images_df.iterrows(), total=len(used_images_df), desc="Performing Object Detection"):
        filename = row["Image"]
        img_path = os.path.join(img_folder, filename)

        if os.path.exists(img_path):
            img = Image.open(img_path)
            img = img.convert('RGB')
            img = img.resize((640, 640))
            img_tensor = to_tensor(img)
            img_tensor = img_tensor.unsqueeze(0)

            # Extract text from the image using Tesseract OCR
            #text = extract_text(img_path)

            cleaned_image_final = remove_captions(img, text_detection_pipeline, text_removal_offset)

           
            if cleaned_image_final.ndim == 3 and cleaned_image_final.shape[-1] == 3:
              cleaned_image_gray = cv2.cvtColor(np.array(cleaned_image_final), cv2.COLOR_RGB2GRAY)
            else:
              cleaned_image_gray = cleaned_image_final


            
            results = model(img_tensor)
            results_np = non_max_suppression(results, conf_thres=0.5, iou_thres=0.5)[0]

            if results_np is not None:
                results_np[:, :4] = scale_boxes(img.size, results_np[:, :4], img.size).round()

                detected_objects = []
                for detection in results_np:
                    class_id = int(detection[5])
                    object_class = model.names[class_id]
                    detected_objects.append(object_class)

                
                cleaned_image_save_path = os.path.join(cleaned_image_path, filename)
                Image.fromarray(cleaned_image_gray).save(cleaned_image_save_path)

                results_list.append({
                    "Sr. No": idx + 1,
                    "Image name": filename,
                    "Detected Categories": detected_objects,
                    "Cleaned Image Path": cleaned_image_save_path
                })

    return results_list

def generate_overall_object_distribution(results_list, output_csv_path):
    all_detected_objects = [obj for result in results_list for obj in result["Detected Categories"]]
    category_counts = pd.Series(all_detected_objects).value_counts().reset_index()
    category_counts.columns = ["Category", "Frequency"]
    category_counts.to_csv(output_csv_path, index=False)

def generate_object_detection_after_caption_removal(results_list, output_csv_path):
    df = pd.DataFrame(results_list)
    df.to_csv(output_csv_path, index=False)

def generate_category_count_after_caption_removal(results_list, output_csv_path):
    df = pd.DataFrame(results_list)
    category_count_dict = {}

    for _, row in tqdm(df.iterrows(), total=len(df), desc="Generating Category Count CSV"):
        categories = row["Detected Categories"]
        image_name = row["Image name"]

        for category in categories:
            if category not in category_count_dict:
                category_count_dict[category] = {}
            category_count_dict[category][image_name] = category_count_dict[category].get(image_name, 0) + 1

    category_count_df = pd.DataFrame(category_count_dict).T.reset_index()
    category_count_df.rename(columns={"index": "Image name"}, inplace=True)
    category_count_df.index.name = "Sr no"
    category_count_df.to_csv(output_csv_path)

def generate_object_detection_with_text_removal_csv(results_list, output_csv_path):
    df = pd.DataFrame(results_list)
    df.to_csv(output_csv_path, index=False)

img_folder = "/content/hateful_memes/img"
weights_path = "/content/yolov5s.pt"
used_images_csv = "/content/used_images_list.csv"

# Step 1: Generate CSV for images after caption removal with text removal
cleaned_image_path = "/content/cleaned_images_with_text_removal"
results_list_with_text_removal = perform_object_detection_with_text_removal(img_folder, weights_path, used_images_csv, cleaned_image_path)

# Step 2: Generate CSV for overall object distribution
overall_object_distribution_path = "/content/overall_object_distribution_with_text_removal2.csv"
generate_overall_object_distribution(results_list_with_text_removal, overall_object_distribution_path)

# Step 3: Generate CSV for images after caption removal with detected objects and confidence
object_detection_with_text_removal_path = "/content/object_detection_with_text_removal2.csv"
generate_object_detection_after_caption_removal(results_list_with_text_removal, object_detection_with_text_removal_path)

# Step 4: Generate CSV for category count after caption removal
category_count_with_text_removal_path = "/content/category_count_with_text_removal2.csv"
generate_category_count_after_caption_removal(results_list_with_text_removal, category_count_with_text_removal_path)

# Step 5: Generate CSV for object detection with text removal
object_detection_with_text_removal_csv_path = "/content/object_detection_with_text_removal_results2.csv"
generate_object_detection_with_text_removal_csv(results_list_with_text_removal, object_detection_with_text_removal_csv_path)
```
